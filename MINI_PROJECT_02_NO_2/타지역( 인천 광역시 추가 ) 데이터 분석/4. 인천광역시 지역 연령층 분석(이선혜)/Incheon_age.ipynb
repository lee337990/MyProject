```python
# ## Series값 두 개 합치기 
many_01 = age_michu_01_S + age_namdong_01_S
type(many_01)

many_12 = age_michu_12_S + age_namdong_12_S
```

```python
manyDF = pd.DataFrame([many_12, many_01])

manyDF
```

```python
littleDF = pd.DataFrame([age_bupeong_12_S, age_bupeong_01_S])

littleDF
```

```python
# 변동률 계산 ( 미추홀구 + 남동구 : 승하차 건수 many )
many_percentage = ((manyDF.iloc[0, 0:] - manyDF.iloc[1, 0:]) / manyDF.iloc[1, 0:]) * 100

# 변동률 계산 ( 부평구 : 승하차 건수 little )
little_percentage = ((littleDF.iloc[0, 0:] - littleDF.iloc[1, 0:]) / littleDF.iloc[1, 0:]) * 100

# 결과 출력
print("롯데백화점 주변 변동률:\n", many_percentage)
print('\n\n')
print("굴포천역 주변 상승률:\n", little_percentage)
```
