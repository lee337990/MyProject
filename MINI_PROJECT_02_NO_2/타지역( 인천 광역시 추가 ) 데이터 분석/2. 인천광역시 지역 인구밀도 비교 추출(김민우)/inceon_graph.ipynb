```python
## [1-1] 모듈 로딩
import pandas as pd                        # 데이터 분석 및 전처리 모듈
import matplotlib.pyplot as plt            # 시각화 모듈
from matplotlib import rc
import platform                             # 한글 깨짐을 막기 위한 한글 폰트 교체를 위한 모듈
if platform.system() == 'Windows':  # Windows
    rc('font', family='Malgun Gothic')


file_path_ipynb = './incheon_people.ipynb'



import nbformat

with open(file_path_ipynb, 'r', encoding='utf-8') as file:
    notebook = nbformat.read(file, as_version=4)


code_cells = [cell for cell in notebook.cells if cell.cell_type == 'code']
namespace = {}

for cell in code_cells:
    try:
        exec(cell.source, namespace)
    except Exception as e:
        pass


datadf = namespace.get('datadf', None)


if datadf is not None:
    print("DataFrame loaded successfully!")
else:
    print("Could not locate 'datadf' in the provided ipynb file.")


datadf








fig, ax = plt.subplots(figsize=(10, 6))


datadf['인구밀도'].plot(kind='bar', ax=ax)


ax.set_ylabel('인구밀도 (명/km²)', fontsize=14)
ax.set_xlabel('')  
ax.set_title('인천 주요 행정구역 인구밀도 비교', fontsize=16)
ax.set_xticklabels(datadf.index, rotation=45, ha='right', fontsize=12)


plt.tight_layout()
plt.show()
```
