<hr> ★★★ 4개의 그래프 한꺼번에 그리기 ★★★ <hr>

```python
years = list(range(2006, 2025))
years1=list(range(2006,2023))

# 그래프 설정 (2x2 레이아웃)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# x축 범위 설정
years_temper = years
years_hot = years
years_rain_hum = years1

# 1. 평균 기온 그래프
axes[0, 0].plot(years_temper, daegu_temper_group["평균기온(℃)"], marker='^', linestyle='-', color='blue', label="평균기온")
axes[0, 0].set_title("년도별 대구 평균 기온 (06년~24년)")
axes[0, 0].set_xlabel("년도")
axes[0, 0].set_ylabel("평균기온(℃)")
axes[0, 0].set_xticks(years_temper)
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].legend()
axes[0, 0].grid(True)

# 2. 폭염 지수 그래프
axes[0, 1].plot(years_hot, daegu_hot_group["연합계"], marker='^', linestyle='-', color='red', label="폭염지수")
axes[0, 1].set_title("년도별 대구 폭염 지수 (06년~24년)")
axes[0, 1].set_xlabel("년도")
axes[0, 1].set_ylabel("폭염지수")
axes[0, 1].set_xticks(years_hot)
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].legend()
axes[0, 1].grid(True)

# 3. 강수량 그래프
axes[1, 0].plot(years_rain_hum, daegu_rain_group["강수량합(mm)"], marker='o', linestyle='-', color='skyblue', label="년도별 강수량 합")
axes[1, 0].set_title("년도별 대구 강수량 합 (06년~22년)")
axes[1, 0].set_xlabel("년도")
axes[1, 0].set_ylabel("강수량 합 (mm)")
axes[1, 0].set_xticks(years_rain_hum)
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].legend()
axes[1, 0].grid(True)

# 4. 평균 상대습도 그래프
axes[1, 1].plot(years_rain_hum, daegu_hum_group["평균상대습도(%)"], marker='o', linestyle='-', color='salmon', label="년도별 평균상대습도")
axes[1, 1].set_title("년도별 대구 평균 상대습도 (06년~22년)")
axes[1, 1].set_xlabel("년도")
axes[1, 1].set_ylabel("평균상대습도(%)")
axes[1, 1].set_xticks(years_rain_hum)
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].legend()
axes[1, 1].grid(True)

# 레이아웃 조정 및 표시
plt.tight_layout()
plt.show()
```

[5] 기온과 폭염지수 상관관계 그래프 <hr>

```python
# x축 범위 설정
years = list(range(2006, 2025))

# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(12, 6))



# 폭염 지수 그래프 
ax1.set_xlabel("년도")
ax1.set_ylabel("폭염 지수", color='r')
ax1.bar(daegu_hot_group["연도"], daegu_hot_group["연합계"], color='pink', label="폭염 지수", zorder=1, alpha=0.8)
ax1.tick_params(axis='y', labelcolor='r')
ax1.set_xticks(years)
plt.legend()

# y의 이차 축 만들기 
ax2 = ax1.twinx() 


# 대구 평균 기온 그래프
# ax2.set_xlabel("년도")
ax2.set_ylabel("평균 기온 (℃)", color='b')
ax2.plot(daegu_temper_group["년"], daegu_temper_group["평균기온(℃)"], marker='^', linestyle='-', color='b', label="평균 기온",zorder=3 )
ax2.tick_params(axis='y', labelcolor='b')
ax2.set_xticks(years)
plt.legend()

# 제목 및 범례 설정
plt.title("년도별 대구 평균 기온 및 폭염 지수 (06년~23년)")
fig.tight_layout()  # 겹치지 않게 배치하기 
plt.grid(True, zorder=0)
plt.show()
```

```python
# 기온과 폭염 데이터만 선택
df_corr = daegu_temper_group["평균기온(℃)"].corr(daegu_hot_group["연합계"])

# 결과 출력
print(f"📌 기온 & 폭염의 상관계수: {df_corr:.4f}")
```

====> 상관관계 결과 0.2215의 결과가 나옴  
- 기온과 폭염 일수는 **약한 양의 상관관계(0.2215)**를 가지며, 기온이 높을수록 폭염이 늘어나는 경향이 있긴 하지만 매우 강한 연관성은 아님.
- 폭염 발생은 단순한 기온 변화만으로 결정되지 않으며, 다른 기상 요인(습도, 강수량, 바람, 일조량 등)이 영향을 미칠 가능성이 큼.
- 상관계수가 낮은 이유를 확인하려면 기온과 폭염 외에도 다른 기상 요소들과의 관계도 분석해보는 것이 필요함.

[6] 기온과 강수량의 상관관계 <hr>

```python
## 기온과 강수량은 데이터양의 차이가 나므로 데이터 갯수를 맞춰야함 2022년까지
daegu_temper_group2022=daegu_temper_group[daegu_temper_group["년"] <=2022]
daegu_temper_group2022
```

```python
## 온도 데이터 22년까지 : daegu_temper_group2022['평균기온(℃)']
## 강수량 데이터 22년까지 : daegu_rain_group['강수량합(mm)']
```

```python
## 그래프 그리기 
## 2006년부터 2022년까지
years = list(range(2006, 2023))  

# 그래프 크기 설정
fig, ax1 = plt.subplots(figsize=(13, 6))


# 첫 번째 Y축 (강수량)
ax1.set_xlabel("년도")
ax1.set_ylabel("강수량 (mm)", color='skyblue')
ax1.bar(years, daegu_rain_group['강수량합(mm)'] , color='skyblue', label="강수량(mm)",zorder=1, alpha=0.8)
ax1.set_xticks(years)
ax1.tick_params(axis='y', labelcolor='skyblue')


# Y의 이차축 만들기
ax2 = ax1.twinx()  


# 두 번째 Y축 (기온)
ax2.set_ylabel("평균 기온 (℃)", color='blue')
ax2.plot(years, daegu_temper_group2022["평균기온(℃)"], 
         marker='o', linestyle='-', color='blue', label="평균 기온(℃)", zorder=3)
ax2.tick_params(axis='y', labelcolor='b')
ax2.set_xticks(years)
plt.legend()


# 제목 및 범례 설정
plt.title("대구 기온 & 강수량 비교 (06년~22년)")
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
fig.tight_layout()  # 겹치지 않게 배치하기 
plt.grid(True, zorder=0)

# 그래프 표시
plt.show()
```

```python
## 상관계수 계산
df_corr_2 = daegu_temper_group2022["평균기온(℃)"].corr(daegu_rain_group['강수량합(mm)'])
## 결과 출력
print(f"📌기온 & 강수량의 상관계수: {df_corr_2:.4f}")
```

====> 상관관계 결과 -0.4612의 결과가 나옴  
- 기온과 강수량은 **중간 정도의 음의 상관관계(-0.4612)**를 가짐.  
- 기온이 상승할수록 강수량이 감소하는 경향이 있지만, 강수량은 다른 기상 요인의 영향을 함께 받음.  
- 더 정확한 해석을 위해 계절별 패턴, 습도, 기압 등의 추가 분석이 필요함.

[7] 기온과 습도의 상관관계 그래프 <hr>

```python
## 온도 데이터 22년까지 : daegu_temper_group2022['평균기온(℃)']
## 습도 데이터 22년까지 : daegu_hum_group['평균상대습도(%)']
```

```python
## x축 범위 설정 : 2006년부터 2022년까지
years = list(range(2006, 2023))  

# 그래프 크기 설정
fig, ax1 = plt.subplots(figsize=(12, 6))

# 첫번째 Y축 (강수량)
ax1.set_xlabel("년도")
ax1.set_ylabel("평균상대습도(%)", color='salmon')
ax1.bar(years,daegu_hum_group['평균상대습도(%)'] ,color='salmon', label="평균상대습도(%)", zorder=1, alpha=0.5)
ax1.tick_params(axis='y', labelcolor='salmon')
ax1.set_xticks(years)
plt.legend()


# Y의 이차 축 만들기
ax2 = ax1.twinx()  


# 두 번째 Y축 (기온)
ax2.set_ylabel("평균 기온 (℃)", color='blue')
ax2.plot(years, daegu_temper_group2022["평균기온(℃)"], marker='o', linestyle='-', color='blue', label="평균 기온(℃)")
ax2.tick_params(axis='y', labelcolor='blue')
ax1.set_xticks(years)
plt.legend()



# 그래프 제목 및  범례 설정
plt.title("대구 기온 & 평균상대습도(%) 비교 (06년~22년)")
plt.grid(True, zorder=0)
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
plt.show()
```

====> 상관관계 결과 0.2595의 결과가 나옴 
- 기온과 습도는 **약한 양의 상관관계 (0.2595)**를 가짐.  
- 기온이 상승할수록 습도가 다소 증가하는 경향이 있지만, 강한 연관성은 없음.  
- 습도는 기온뿐만 아니라 강수량, 대기압, 바람, 계절적 요인 등 여러 변수에 영향을 받음.

```python
# x축 데이터 설정 (2006년~2022년)
years = list(range(2006, 2023))
bar_width = 0.3  # 막대 너비 설정
x_indices = range(len(years))  # x축 위치

# 데이터가 2006~2022년 범위를 포함하는지 필터링
daegu_hot_group = daegu_hot_group[daegu_hot_group["연도"].between(2006, 2022)]
daegu_hum_group = daegu_hum_group[daegu_hum_group["년도"].between(2006, 2022)]
daegu_rain_group = daegu_rain_group[daegu_rain_group["년도"].between(2006, 2022)]

# 길이 확인 (일치하지 않으면 오류 발생 가능)
assert len(daegu_hot_group) == len(years), "폭염지수 데이터 개수가 연도와 다름"
assert len(daegu_hum_group) == len(years), "습도 데이터 개수가 연도와 다름"
assert len(daegu_rain_group) == len(years), "강수량 데이터 개수가 연도와 다름"

# 그래프 설정
fig, ax1 = plt.subplots(figsize=(12, 6))

# 막대 그래프 (폭염지수 & 평균 상대습도, 왼쪽 y축)
ax1.bar([x - bar_width for x in x_indices], daegu_hot_group["연합계"], width=bar_width, color='coral', label="폭염지수")
ax1.bar([x for x in x_indices], daegu_hum_group["평균상대습도(%)"], width=bar_width, color='lightblue', label="평균 상대습도")
ax1.set_xlabel("년도")
ax1.set_ylabel("폭염지수 / 평균 상대습도 (%)")
ax1.set_xticks(x_indices)
ax1.set_xticklabels(years, rotation=45)
ax1.tick_params(axis='y', labelcolor='black')

# 막대 그래프 (강수량, 오른쪽 y축)
ax2 = ax1.twinx()
ax2.bar([x + bar_width for x in x_indices], daegu_rain_group["강수량합(mm)"], width=bar_width, color='plum', label="강수량(mm)", alpha=0.7)
ax2.set_ylabel("강수량 (mm)", color='green')
ax2.tick_params(axis='y', labelcolor='green')

# 타이틀 및 범례 설정
plt.title("폭염지수 & 상대습도 & 강수량 (2006년~2022년)")
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

# 그래프 표시
plt.grid(True)
plt.show()
```
